```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
x=np.arange(-10,10.1,0.5),
y=np.arange(-10,10.1,0.5),
```

```python
print(x)
print(y)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
Z=X**2+Y**2
```

```python
fig=plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
surf=ax.contour3D(X, Y, Z, 200, cmap=plt.cm.rainbow)
ax.contourf(X, Y, Z, zdir="z",offset=0,cmap=plt.cm.rainbow)
fig.colorbar(surf,shrink=0.5,aspect=5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
